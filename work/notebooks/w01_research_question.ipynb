## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*
Here are real numbers from the starter dataset that support this lane.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)

print(df.shape)
declining_rate = df["is_declining_label"].mean()
median_impressions = df["impressions_90d"].median()
stale_and_visible = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).sum()

print(f"Share of pages currently declining: {declining_rate:.3f}")
print(f"Median 90-day impressions: {median_impressions:.0f}")
print(f"Pages that are stale AND still visible: {stale_and_visible} out of {len(df)}")
```

Of the 30,000 pages in the starter dataset, 54.2% are currently labeled as declining, and the median page still receives 731 impressions over 90 days — so this is a dataset with real visibility at stake, not abandoned content. However, only 17 pages meet the strict "stale AND visible" rule from Week 1, showing that a single narrow baseline won't surface enough candidates on its own. This confirms Lane 2 is worth pursuing, but also tells me my baseline will need more than one rule (e.g. combining "declining with demand" and "page-one decay risk" from the lane guide) to build a useful, appropriately-sized review queue.

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

What this work can say: Based on the starter dataset, I can identify pages that are observed to be declining, stale, or under-visible using transparent, explainable rules and simple models — and I can rank them so a reviewer knows where to look first. Any relationship I find between signals (like staleness, impressions, or position) and decline is associative and directional — it describes a pattern in the data, not a guaranteed cause. My output is decision-support: a ranked list with reasons attached, meant to guide a human reviewer's limited time, not replace their judgment.

What this work can never say: I cannot claim that refreshing a page causes it to recover — proving that would require an actual experiment (e.g. refreshing some pages and comparing them to similar pages left untouched), which this dataset does not provide. I also cannot claim to have discovered anything about how Google's search algorithm actually works — I'm only observing search performance signals, not the ranking system itself. A high score from my model means "worth reviewing first," never "guaranteed to improve."
